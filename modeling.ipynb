```python
import pandas as pd
```

```python
movies = pd.read_csv('csv_data/10000 Movies Data.csv')
movies.rename(columns={'Movie_id':'id','Keywords':'keywords','Genres':'genres'},inplace=True)
movies.drop_duplicates(subset=['title'],keep='first',inplace=True)
movies.head()
```

```python
cradits = pd.read_csv('csv_data/10000 Credits Data.csv')
cradits.rename(columns={'Movie_id':'movie_id','Cast':'cast','Crew':'crew'},inplace=True)
cradits.head()
```

```python
cradits.drop('title',axis=1)
```

## merging data

```python
movies = movies.merge(cradits,left_on='id',right_on='movie_id', suffixes=('x',''))
movies.shape
```

```python
movies_data = movies
movies.sample(10,random_state=200)
```

## columns to keep
- genres
- title
- overview
- keywords
- id

```python
colums_to_keep = ['id','title','keywords','genres','cast','overview','crew']
movies = movies[colums_to_keep]
movies
```

```python
movies['keywords'][0]
```

## handling keywords and geners

```python
import ast

def convert_tags_to_list(obj):
    li = []
    for i in ast.literal_eval(obj):
        li.append(i['name'])
    return li
```

```python
movies['keywords'] = movies['keywords'].apply(convert_tags_to_list)
movies['genres'] = movies['genres'].apply(convert_tags_to_list)
```

```python
movies.head()
```

```python
movies['cast'][0]
```

## handling cast
_top 3_

```python
def convert_cast_to_list(obj):
    cnt=0
    li = []
    for i in ast.literal_eval(obj):
        if cnt!=3:
            li.append(i['name'])
            cnt+=1
        else:
            return li
    return li
```

```python
movies['cast'] = movies['cast'].apply(convert_cast_to_list)
movies['cast']
```

```python
movies['crew'][0]
```

## handling crew as diractor

```python
def convert_crew_to_diractor(obj):
    li = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            li.append(i['name'])
            return li        
```

```python
movies['crew'] = movies['crew'].apply(convert_crew_to_diractor)
```

```python
movies
```

```python
movies.fillna("",inplace=True)
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies
```

### replacing the space with _

```python
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","_") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","_") for i in x])
# movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","_") for i in x])
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","_") for i in x])
movies.head()
```

## merging all in single dataset as tags

```python
tags = movies['keywords'] + movies['genres'] + movies['overview'] + movies['crew'] + movies['cast']
tags
```

```python
data = movies[['id','title']]
data['tags'] = tags
data.sample(n=5,random_state=200)
```

```python
data['tags'] = data['tags'].apply(lambda x:" ".join(x))
data
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def get_stem(text):
    li = []
    for i in text.split():
        li.append(ps.stem(i))
    return " ".join(li)    
```

```python
data['tags']= data['tags'].apply(get_stem)
```

## build vector

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
vectors = cv.fit_transform(data['tags']).toarray()
```

## get cosine similarity distances

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
similarity.shape
```

```python
sorted(enumerate(similarity[0]),reverse=True,key=lambda x: x[1])[:6]
```

## recommand function

```python
def recommend(movie):
    movie_index = data[data['title']==movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(enumerate(distances),reverse=True,key=lambda x: x[1])[1:20]

    for i in movie_list:
        print(data.iloc[i[0]].title)
    
```

```python
recommend('Spectre')
```
